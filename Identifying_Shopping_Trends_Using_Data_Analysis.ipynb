```python
import matplotlib.pyplot as plt
import pandas as pd

# ... (your existing code) ...

# Customize the box plot
boxprops = dict(linestyle='-', linewidth=2, color='blue')  # Customize box properties
flierprops = dict(marker='o', markerfacecolor='red', markersize=8, linestyle='none')  # Customize outlier properties
medianprops = dict(linestyle='-', linewidth=2, color='green')  # Customize median line properties
whiskerprops = dict(linestyle='--', linewidth=1.5, color='orange')  # Customize whisker properties
capprops = dict(linestyle='-', linewidth=2, color='black')  # Customize cap properties

df.boxplot(column="Review Rating", by="Gender",
           boxprops=boxprops, flierprops=flierprops, medianprops=medianprops,
           whiskerprops=whiskerprops, capprops=capprops, patch_artist=True)

plt.xlabel("Gender", fontsize=12)
plt.ylabel("Review Rating", fontsize=12)
plt.title("Distribution of Review Ratings by Gender", fontsize=14, fontweight='bold')
plt.suptitle('')  # Remove the default suptitle
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ... (your existing code) ...

# Customize the histogram
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
n, bins, patches = plt.hist(df['Purchase Amount (USD)'], bins=10, edgecolor='black', color='skyblue')

# Add a gradient color effect to the bars
cm = plt.cm.get_cmap('viridis')  # Choose a colormap
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

plt.xlabel("USD", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Histogram of Purchase Amount", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid to the y-axis

plt.show()
```
